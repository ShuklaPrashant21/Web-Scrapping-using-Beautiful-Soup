###USE REGULAR EXPRESSION
Regular Expression is used to text processing where you find out patterns that is basically pattern matching where you go for regex is used in javascript and many more.

```
import re
list_rows = []
for row in rows:
  cells = row.find_all('td')
  str_cells = str(cells)
  clean = re.compile('<.*?>')
  clean2 = (re.sub(clean,'',str_cells))
  list_rows.append(clean2)
  print(clean2)
```
